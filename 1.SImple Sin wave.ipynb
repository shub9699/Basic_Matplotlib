```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

#### creating an array from 0 to pi with intervals of 0.5

```python
x = np.arange(0, math.pi*2, 0.05)
```

```python
x
```

#### applying sin function to each value in the x

```python
y = np.sin(x)
y
```

#### Basic x and y plot

```python
plt.plot(x,y)
```

### giving labels

```python
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title("sine wave")
plt.show()
```
