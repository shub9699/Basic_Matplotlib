ALSO IN DIFFERENT COLOUR

```python
x = linspace(-3, 3, 30)
y = x**2
plot(x, y, 'r+')
show()
```

```python
x = linspace(-3, 3, 30)
y = x**2
plot(x, y, 'r.')
show()
```

```python
x = linspace(-3, 3, 30)
y = x**2
plot(x, y, 'b*')
show()
```

The "r.,", "b*" is the notation which means red color with dot symbol, blue with star
